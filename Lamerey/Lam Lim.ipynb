```python
# Логистическая
import matplotlib.pyplot as plt
from numpy import *
def Lamer(a,k,H): # коэф А, Х0, кол-во шагов
    x1=3  # До какого момента делаем вычисление
    def ff(x):   # это основная функция
        return a*x*(1-x)
    b=a*x1*(1-x1)/x1 
    def fl(x):  # уравнение x=y
            return x
    def ft(x):
        return a*(a*x*(1-x))*(1-a*x*(1-x))
    x=k   # X0
    y=0
    Y=[]
    X=[]
    for i in arange(1,H,1):  # от 1 до H с шагом 1
            X.append(x)
            Y.append(y)
            y=ff(x)
          #  print('X= ',x,' Y= ',y)
            X.append(x)
            Y.append(y)
            x=y/b
    plt.title('Паутинная диаграмма логистической \n функции Ax(1-x) при A = '+str(a))
    plt.plot(X,Y,'r')  # Ламерей
    x1=arange(0,1,0.01) # от 0 до 1 с шагом 0.01
    y1=[ff(x) for x in x1]
    y2=[fl(x) for x in x1]
    y3=[ft(x) for x in x1]
    plt.plot(x1,y1,'b') # основная функция
    plt.plot(x1,y2,'g') # график х=у
    plt.plot(x1,y3,'k')
    plt.grid(True)
    plt.show()
```
